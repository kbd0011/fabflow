# 03 · Dashboard preview\n\nThe Streamlit dashboard (`make dashboard`) reads the same DuckDB warehouse. Here we render its core charts inline so you can preview them without launching Streamlit.

```python
import duckdb, plotly.express as px\nfrom fabflow.config import load_config\nfrom fabflow.analysis.kpis import utilization\n\ncfg = load_config('../configs/config.yaml')\ndb = str(cfg.output.warehouse)  # built in notebook 02\ncon = duckdb.connect(db, read_only=True)\nlot = con.execute('SELECT scenario, cycle_time FROM fact_lot').df()\ncon.close()\npx.histogram(lot, x='cycle_time', color='scenario', barmode='overlay', nbins=50, histnorm='probability density', opacity=0.55, title='Cycle-time distribution')
```

```python
u = utilization(db, cfg.sim.horizon_hours, cfg.sim.warmup_hours)\nfig = px.bar(u, x='name', y='utilization', color='scenario', barmode='group', title='Tool-group utilization')\nfig.add_hline(y=1.0, line_dash='dash', line_color='red')\nfig
```
